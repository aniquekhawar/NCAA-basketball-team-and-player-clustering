# Import Libraries

```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

# Import Teams Dataset

> The Teams dataset needs columns to have specific names for use in the BasketballAnalyzeR package in R. The only thing this dataset is missing a plus/minus statistic, but we opted not to use it in our final analysis. The package also expects team stats to be separated from opponent's stats.

```python
df = pd.read_csv('data/ncaa_teams_2010_to_2020.csv')
obox_df = df.copy()
df.head()
```

## Creating Tbox

```python
tbox_cols = {'abbreviation': 'Team',
                'two_point_field_goal_attempts': 'P2A',
                'two_point_field_goals': 'P2M',
                'three_point_field_goal_attempts': 'P3A',
                'three_point_field_goals': 'P3M',
                'free_throw_attempts': 'FTA',
                'free_throws': 'FTM',
                'offensive_rebounds': 'OREB',
                'defensive_rebounds': 'DREB',
                'turnovers': 'TOV',
                'minutes_played': 'MIN',
                'games_played': 'GP',
                'points': 'PTS',
                'wins': 'W',
                'losses': 'L',
                'two_point_field_goal_percentage': 'P2p',
                'three_point_field_goal_percentage': 'P3p',
                'free_throw_percentage': 'FTp',
                'assists': 'AST',
                'steals': 'STL',
                'blocks': 'BLK',
                'personal_fouls': 'PF'}
df.rename(columns = tbox_cols, inplace = True)
tbox_values = list(tbox_cols.values())
tbox = df[tbox_values]
tbox
```

## Creating Obox

```python
obox_cols = {'abbreviation': 'Team',
                'opp_two_point_field_goal_attempts': 'P2A',
                'opp_two_point_field_goals': 'P2M',
                'opp_three_point_field_goal_attempts': 'P3A',
                'opp_three_point_field_goals': 'P3M',
                'opp_free_throw_attempts': 'FTA',
                'opp_free_throws': 'FTM',
                'opp_offensive_rebounds': 'OREB',
                'opp_defensive_rebounds': 'DREB',
                'opp_turnovers': 'TOV',
                'minutes_played': 'MIN',
                'games_played': 'GP',
                'opp_points': 'PTS',
                'wins': 'W',
                'losses': 'L',
                'opp_two_point_field_goal_percentage': 'P2p',
                'opp_three_point_field_goal_percentage': 'P3p',
                'opp_free_throw_percentage': 'FTp',
                'opp_assists': 'AST',
                'opp_steals': 'STL',
                'opp_blocks': 'BLK',
                'opp_personal_fouls': 'PF'}
obox_df.rename(columns = obox_cols, inplace = True)
obox = obox_df[tbox_values]
obox
```

## Exporting final Tbox and Obox files

```python
tbox.to_csv('data/teams_tbox.csv', index = False)
obox.to_csv('data/teams_obox.csv', index = False)
```

# Import Players Dataset
> Clustering NCAA players may be a little more challenging, so we have to decide whether we want to take career stats or stats from a player's latest season (we should note some players in this dataset aren't drafted into the NBA). Regardless, because this is a play style analysis of players on Kentucky and Duke, we should be fair to analyze all players.

```python
df2 = pd.read_csv('data/players_duke_kentucky_2010_to_2020.csv', index_col = 'id')
df2.head()
```

```python
pbox_cols = {'team_abbreviation': 'Team',
                'name': 'Player',
                'two_point_attempts': 'P2A',
                'two_pointers': 'P2M',
                'three_point_attempts': 'P3A',
                'three_pointers': 'P3M',
                'free_throw_attempts': 'FTA',
                'free_throws': 'FTM',
                'offensive_rebounds': 'OREB',
                'defensive_rebounds': 'DREB',
                'turnovers': 'TOV',
                'minutes_played': 'MIN',
                'games_played': 'GP',
                'points': 'PTS',
                'two_point_percentage': 'P2p',
                'three_point_percentage': 'P3p',
                'free_throw_percentage': 'FTp',
                'assists': 'AST',
                'steals': 'STL',
                'blocks': 'BLK',
                'personal_fouls': 'PF',
                'box_plus_minus': 'PM',
                'year': 'Year'}
df2.rename(columns = pbox_cols, inplace = True)
pbox_values = list(pbox_cols.values())
pbox = df2[pbox_values]
pbox
```

```python
pbox = pbox[(pbox['Team'] == 'kentucky') | (pbox['Team'] == 'duke')]
```

## Exporting final Pbox file.

```python
pbox.to_csv('data/pbox_duke_kentucky.csv', index = False)
```
